# Tarea 5.6

Un cierto sensor se deja abierto por un segundo. El sensor consiste en $r$ pixeles y en promedio el sensor es golpeado por $2$ partículas por segundo, tal que la probabilidad de detectar una partícula en un pixel dado es de $2/r$ (para $r \gg 2$). Grafique la distribución de probabilidad de observar $n$ partículas para un pixel dado, en función de $n$, para diferentes valores de $r$:

* Usando la distribución binomial.

* ¿Cuándo es posible usar la distribución de Poisson?

# Respuesta:

Se nos entrega que la probabilidad del sensor para detectar particulas a lo largo de 1 segundo es $2/r$ (para $r \gg 2$)

Tambien vimos en clase la fórmula de la Distribución de Poisson en su forma binomial, que es: 

$$
P(n; r, \lambda/r) = \frac{\lambda^n}{r^n}\left(1 - \frac{\lambda^n}{r^n}\right)^{r - n}\frac{r!}{n!(r-n)!}
$$

Para simular este experimento podemos ocupar las funciones vistas en clases para obtener la probanilidades o para realizar el experimiento una cantidad de veces arbitrarias 

```python
from scipy.special import comb
def poisson(num_particulas, num_pixeles):
  '''
  expresion que nos permite calcular la distribucion de Poisson en funcion de
  el numero de particulas, el numero de pixeles y lamda que es la probabilidad 
  de impoacto de la particula en un sensor a lo largo de 1 segundo 
  '''
  lamda = (2/num_pixeles)
  return comb(num_pixeles, num_particulas)*lamda**(num_particulas)*(1-lamda**(num_particulas))**(num_pixeles - num_particulas)
```

```python
poisson(5,20)
```

Ahora que ya tenemos una funsión que ejecuta la distribución de Poisson crearemos una funsión que calcule las probabilidades 

```python
def probs(num_particulas, num_pixeles):
    prob_experimento= 2/num_pixeles
    return [poisson(num_particulas, num_pixeles) for num_particulas in range(num_pixeles+1)]
```

Ahora graficaremnos distintos casos variando tanto como la cantidad de pixeles como el numero de particulas

```python
fig = plt.figure(figsize=(8,12), dpi= 100)

ax1 = fig.add_subplot(4,2,1)
ax2 = fig.add_subplot(4,2,2)
ax3 = fig.add_subplot(4,2,3)
ax4 = fig.add_subplot(4,2,4)
ax5 = fig.add_subplot(4,2,5)
ax6 = fig.add_subplot(4,2,6)
ax7 = fig.add_subplot(4,2,7)
ax8 = fig.add_subplot(4,2,8)

ax1.bar(range(5+1), probs(0.2, 5))
ax2.bar(range(5+1), probs(0.5, 5))
ax3.bar(range(10+1), probs(1, 10))
ax4.bar(range(10+1), probs(2, 10))
ax5.bar(range(15+1), probs(5, 15))
ax6.bar(range(25+1), probs(10, 25))
ax7.bar(range(40+1), probs(20, 40))
ax8.bar(range(80+1), probs(50, 80))
        
ax1.set_title('$\lambda = 0.2$')
ax2.set_title('$\lambda = 0.5$')
ax3.set_title('$\lambda = 1$')
ax4.set_title('$\lambda = 2$')
ax5.set_title('$\lambda = 5$')
ax6.set_title('$\lambda = 10$')
ax7.set_title('$\lambda = 20$')
ax8.set_title('$\lambda = 50$')

fig.tight_layout(pad=2.0)


fig.show()
```
